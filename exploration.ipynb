```python
def myhash(a: int) -> int:
    """A simple 32-bit hash function"""
    fns = {
        "+": lambda x, y: x + y,
        "^": lambda x, y: x ^ y,
        "<<": lambda x, y: x << y,
        ">>": lambda x, y: x >> y,
    }

    def r(x):
        return x % (2**32)

    for i, (op1, val1, op2, op3, val3) in enumerate(HASH_STAGES):

        a = r(fns[op2](r(fns[op1](a, val1)), r(fns[op3](a, val3))))
        # print(f"After Stage {i}: {bin(a)}, {hex(a)}, {a})")

    return a
```
